```python
from google.colab import files
from google.colab.patches import cv2_imshow
import numpy as np
import cv2 as cv

def process_image_and_display(img, number):
    # Convert the image to grayscale
    gray = cv.cvtColor(img, code=cv.COLOR_BGR2GRAY)

    # Process the image based on the specified number
    if number == 1:
        dil_img = cv.dilate(gray, np.ones((5, 5)))
        blur_img = cv.GaussianBlur(dil_img, (7, 7), 128)
        final = 255 - cv.absdiff(blur_img, gray)

        for i in range(0, img.shape[0]):
            for j in range(0, 100):
                final[i][j] = (final[i][j] / 255) ** (1.4) * 255

    elif number == 2:
        dil_img = cv.dilate(gray, np.ones((19, 19)))
        blur_img = cv.medianBlur(dil_img, 23)
        final = 255 - cv.absdiff(blur_img, gray)

    elif number == 3:
        final = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 3, 2)

    # Convert the data type to uint8 and display the processed image
    final_display = final.astype(np.uint8)
    cv2_imshow(final_display)

# Upload an image
uploaded = files.upload()

# Get the file name
file_name = list(uploaded.keys())[0]

# Read the uploaded image
img = cv.imread(file_name)

# Specify the number for processing (1, 2, or 3)
process_number = 1

# Process and display the image
process_image_and_display(img, process_number)
```
