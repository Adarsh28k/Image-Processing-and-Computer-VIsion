```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
from google.colab import files

uploaded = files.upload()

file_name = list(uploaded.keys())[0]
img = cv.imread(file_name)

img1 = img[150:330, 515:700]
img1 = cv.flip(img1, 1)
img[150:330, 515:700] = img1

img2 = img[370:420, 370:795]
img2 = cv.flip(img2, 0)
img[370:420, 370:795] = img2

img3 = img[0:200, 0:190]

img4 = img[200:400, 0:190]
img4 = cv.flip(img4, 0)
img[200:400, 0:190] = img4
img[0:200, 0:190] = img3

cv2_imshow(img)
cv.waitKey(0)
```
